## 5. Visualizacao das metricas


```python
from pathlib import Path
import csv

csv_path = Path("resultados_clevr_ltn.csv")
rows = list(csv.DictReader(csv_path.open(newline="", encoding="utf-8")))

try:
    import pandas as pd
    display(pd.read_csv(csv_path))
except ImportError:
    for row in rows:
        print(row)
```

## 6. Conferencia das 5 classes


```python
for row in rows:
    counts = {k: row[k] for k in ["n_circle", "n_square", "n_ellipse", "n_rectangle", "n_triangle"]}
    print(f"teste seed={row['seed']}:", counts)
```

## 7. Evidencia das consultas compostas e auditoria XAI


```python
for row in rows:
    print(
        f"seed={row['seed']} | "
        f"q1_gt={row['q1_gt_exists']} conf={float(row['q1_best_witness_conf']):.3f} obj={row['q1_best_witness_obj']} | "
        f"q2_gt={row['q2_gt_exists']} conf={float(row['q2_best_witness_conf']):.3f} obj={row['q2_best_witness_obj']} | "
        f"q3_pares={row['q3_triangle_close_pairs']} violacoes={row['q3_triangle_size_violations']} | "
        f"xai_left={row['xai_leftmost_obj']} right={row['xai_rightmost_obj']} "
        f"conf={float(row['xai_leftmost_rightmost_conf']):.3f} rev={float(row['xai_reverse_conf']):.3f} "
        f"asym={float(row['xai_asymmetry_conf']):.3f}"
    )
```

## 8. Resumo das metricas principais


```python
import statistics
for col in ["satAgg", "accuracy", "precision", "recall", "f1"]:
    values = [float(row[col]) for row in rows]
    print(f"{col:10s} media={statistics.mean(values):.4f} desvio={statistics.pstdev(values):.4f}")
```

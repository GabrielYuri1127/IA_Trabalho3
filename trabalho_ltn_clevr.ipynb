## 5. Visualizacao das metricas


```python
from pathlib import Path
import csv

csv_path = Path("resultados_clevr_ltn.csv")
rows = list(csv.DictReader(csv_path.open(newline="", encoding="utf-8")))

try:
    import pandas as pd
    display(pd.read_csv(csv_path))
except ImportError:
    for row in rows:
        print(row)
```

## 6. Conferencia das 5 classes


```python
for row in rows:
    counts = {k: row[k] for k in ["n_circle", "n_square", "n_cylinder", "n_cone", "n_triangle"]}
    print(f"teste seed={row['seed']}:", counts)
```

## 7. Resumo das metricas principais


```python
import statistics
for col in ["satAgg", "accuracy", "precision", "recall", "f1"]:
    values = [float(row[col]) for row in rows]
    print(f"{col:10s} media={statistics.mean(values):.4f} desvio={statistics.pstdev(values):.4f}")
```
